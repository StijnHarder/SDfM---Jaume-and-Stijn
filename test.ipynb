```python
# merge each df with an left join to only include movieids of which there are details provided within the ratings, links and tags csv  
merged_df = pd.merge(df_movies, df_ratings, on='movieId', how='left')
merged_df = pd.merge(merged_df, df_links, on='movieId', how='left')
merged_df = pd.merge(merged_df, df_tags, on='movieId', how='left')
merged_df.info()
merged_df
```

```python
import re
# extract the four digits between brackets by using a regular expressions pattern to look for the pattern, which will then be extracted by the lamba function
merged_df['year'] = merged_df['title'].apply(lambda x: re.search(r'\((\d{4})\)', x).group(1) if re.search(r'\((\d{4})\)', x) else None)
# cut off last six characters to keep the clean movie title
merged_df['title'] = merged_df['title'].str[:-6]
merged_df
```

```python
merged_df['timestamp_x'] = timestamp_x = pd.to_datetime(merged_df['timestamp_x'], unit='s')
merged_df['timestamp_y'] = timestamp_x = pd.to_datetime(merged_df['timestamp_y'], unit='s')
merged_df.isnull().sum()
merged_df
```

Group by movie id and title to create a dataframe which can be analyzed:

```python
df = merged_df.groupby(['movieId', 'title']).agg({
    'genres': 'first',
    'userId_x': 'count',
    'rating': 'mean',
    'timestamp_x': 'min',
    'imdbId': 'max',
    'tmdbId': 'first',
    'timestamp_y': 'min',
    'year': 'max'
}).reset_index()

df = df.rename(columns={
    'userId_x':'amount_of_ratings',
    'rating':'avg_rating' 
})
df.dtypes
df
```
